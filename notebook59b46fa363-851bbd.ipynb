```python
df = pd.read_csv('/kaggle/input/pythonpro-digital-edu/train.csv')
df.head(20)
```

```python
various = list()
def count_mean_bdata(row):
    if len(str(row['bdate'])) > 5:
        global various
        bd = row['bdate'].split('.')
        time = int(row['graduation']) - int(bd[2])
        if row['graduation'] != 0:
            various.append(time)
            return sum(various) / len(various)
    else:
        return row
        

df.apply(count_mean_bdata, axis=1)
     
```

```python
def fill_age(row):
    if len(str(row['bdate'])) > 5:
        age = str(row['bdate'])
        age = age.split('.')
        row['bdate'] = age[2]
        return row
    elif len(str(row['bdate'])) < 5 and row['graduation'] != 0.0:
        age = int(row['graduation']) -  25
        row['bdate'] = int(age)
        return row
    elif row['bdate'] == 'NaN' and row['graduation'] != 0.0:
        age = int(row['graduation']) -  25
        row['bdate'] = int(age)
        return row
    else:
        return row
df= df.apply(fill_age, axis=1)
```

```python
def fill_education_form(row):
    if row['education_form'] == 'Full-time':
        row['education_form'] = 1
        return row
    elif row['education_form'] == 'Distance Learning':
        row['education_form'] = 2
        return row
    else: 
        row['education_form'] = 3
        return row
df = df.apply(fill_education_form, axis=1)
df.info()
df
```

```python
# 1985
# 1986
df[(df['occupation_type'] == 'work')]['bdate'].mean()
def fill_occupy(row):
    if row['occupation_type'] == 'NaN':
        if row['bdate'] > 1986:
            row['occupation_type'] = 0
            return row
        else:
            row['occupation_type'] = 1
            return row
    else:
        if row['occupation_type'] == 'university':
            row['occupation_type'] = 0
            return row
        else:
            row['occupation_type'] = 1
            return row
df = df.apply(fill_occupy, axis = 1)

```

```python
def fill_career_start(row):
    if row['career_start'] == 'False':
        row['career_start'] = 0
        return row
    else:
        row['career_start'] = 1

        return row
df = df.apply(fill_career_start, axis=1)

```

```python
def fill_g(row):
    if row['education_status'] == 'Alumnus (Specialist)':
        row['education_status'] = 1
        return row
    elif row['education_status'] == 'Student (Specialist)':
        row['education_status'] = 2
        return row
    elif row['education_status'] == "Student (Bachelor's)":
        row['education_status'] = 3
        return row
    elif row['education_status'] == "Alumnus (Bachelor's)":
        row['education_status'] = 4
        return row
    elif row['education_status'] == "Alumnus (Master's)":
        row['education_status'] = 5
        return row
    elif row['education_status'] == 'PhD':
        row['education_status'] = 6
        return row
    elif row['education_status'] == "Student (Master's)":
        row['education_status'] = 7
        return row
    elif row['education_status'] == 'Undergraduate applicant':
        row['education_status'] = 7
        return row
    else:
        row['education_status'] = 7
        return row
df = df.apply(fill_g, axis = 1)
```

```python
df.drop(['langs', ], axis=1, inplace = True)
```

```python
def last_seen(row):
    row['last_seen']=row['last_seen'][:4]
    return row
df = df.apply(last_seen, axis=1)
```

```python
def people_main(row):
    if row['people_main'] == 'False':
        row['people_main'] = 7
        return row
    else:
        row['people_main'] = row['people_main']
        return row
df = df.apply(people_main, axis = 1)
    
```

```python
def life_main(row):
    if row['life_main'] == 'False':
        row['life_main'] = 9
        return row
    else:
        row['life_main'] = row['life_main']
        return row
df = df.apply(life_main, axis = 1)
```

```python
def int_life_main(row):
    if type(row['life_main']) != 'int':
        row['life_main'] = int(row['life_main'])
        return row
    else:
        return row
df = df.apply(int_life_main, axis=1)
```

```python
def int_last_seen(row):
    if type(row['last_seen']) != 'int':
        row['last_seen'] = int(row['last_seen'])
        return row
    else:
        return row
df = df.apply(int_last_seen, axis=1)
```

```python
def int_people_main(row):
    if type(row['people_main']) != 'int':
        row['people_main'] = int(row['people_main'])
        return row
    else:
        return row
df = df.apply(int_people_main, axis=1)
```

```python
from sklearn.model_selection import train_test_split
y = df['result']
x = df.drop('result',axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,
                                                    y,
                                                    test_size=0.25)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)                
x_test = sc.fit_transform(x_test)
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=7)          
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix, accuracy_score
print('Процент правильно предсказать итог:',accuracy_score(y_test,y_pred)*100)
```
